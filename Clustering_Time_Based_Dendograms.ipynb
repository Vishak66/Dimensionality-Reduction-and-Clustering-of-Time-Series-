### Finding the peak widths

```python
temp_mean=[None]*6579
```

```python
for i,j in enumerate(max):      
    temp=[None]*len(j)
    for m,n in enumerate(j):
        if(len(max[i])==len(min[i])):
            temp[m]=(X_max[i][m]+X_min[i][m])/2
        else:
            break
    temp_mean[i]=temp
        
        
```

```python
temp_mean[0:4]
```

```python
peak_width=[None]*6579
```

```python
# To find the nearest value

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

```python
for i,j in enumerate(max):      
    temp=[None]*len(j)
    for m,n in enumerate(j):
        if(len(max[i])==len(min[i])):
            arr=[]
            arr=X_narray[i][n:min[i][m]]
            temp[m]=find_nearest(arr,temp_mean[i][m])      
    peak_width[i]=temp
    
```
